# 1. Conducting the initial study

This is a classic typing experiment followed by a parameter estimation analysis task. A smartwatch company decided to provide the text entry function to their customers. Their solution was straightforward; they directly displayed a qwerty keyboard on the smartwatch. To ensure this is an acceptable design, they invited 50 users to participate in a user study. The only measurement they collected was the error rate, which are shown below (unit: %). 

![keyboard](images/keyboard.jpg)

```python
error_rate = [24.512, 32.195, 25.611, 35.576, 32.158, 18.731, 31.437, 26.643, 45.866, 35.134, 
              30.499, 33.387, 33.262, 26.232, 22.53, 35.449, 29.985, 41.053, 23.28, 23.698, 
              38.832, 23.071, 12.984, 33.73, 33.893, 28.5, 21.867, 35.697, 33.877, 41.395, 
              26.586, 29.913, 22.655, 48.285, 17.953, 32.941, 21.287, 28.368, 32.517, 32.498, 
              23.71, 45.1, 10.119, 7.923, 32.634, 20.555, 25.022, 26.73, 12.627, 29.69]
```

## 1.1 Parameter estimation from a Frequentist's perspective

Is the average of the error rates different from 31%?

```python
print(st.ttest_1samp(error_rate,31))
```

Because the p-value is NOT < 0.05, so we did not reject the null hypothesis (mean = 31%).

## 1.2 Parameter estimation from a Bayesian perspective

With the same data, now let's run the analysis from a Bayesian perspective.

```python
# Step 1. collect some data from users
# Step 2. Parameterize your problem using statistical distributions
# Step 3. Write model in PyMC3, and justify your model structure

with pm.Model() as error_rate_model: 
    
    
    mu_distribution = pm.Uniform('mu_distribution', 20, 45)
    sigma_distribution = pm.Uniform('sigma_distribution', 0, 15)
    like = pm.Normal('likelihood', mu=mu_distribution, sigma=sigma_distribution,
                        observed=error_rate)  

    # Step 4. Inference
    step = pm.Metropolis()
    error_rate_trace = pm.sample(1000, step=step,cores=4)
```

```python
# Step 5. Interpret the results
pm.traceplot(error_rate_trace)
plt.show()
```

```python
pm.plot_posterior(error_rate_trace, 
                  color='#87ceeb', point_estimate='mean', 
                  rope=[30.9, 31.1], ref_val=31)
plt.show()
```

Is 31% within the 94% HDI?

# 2. Conducting the replication studies


Say the company decided to replicate the experiment with less participants (12 ppl) for 2 times just to make sure the results of the previous study is sound. What will they learn from each perpective?

```python
error_rate_2 = [28.07, 28.524, 29.156, 31.094, 30.308, 30.602, 22.325, 26.221, 30.45,  32.576, 31.362, 33.616]

error_rate_3 = [27.837, 27.848, 33.072, 29.379, 24.784, 25.947, 25.053, 26.183, 30.823, 27.506, 20.945, 29.619]
```

## 2.1 Parameter estimation from a Frequentist's perspective

```python
print(st.ttest_1samp(error_rate_2,31))
print(st.ttest_1samp(error_rate_3,31))
```

### What is the conclusion if we simply use "vote-counting?"

Initial study and the second study finds no significance while the third study finds significance.

## 2.2 Meta-analysis from a Frequentist's perspective

```python
print(st.ttest_1samp(error_rate+error_rate_2+ error_rate_3,31))
```

## 2.3 Parameter estimation from a Bayesian perspective

```python
# Step 1. collect some data from users
# Step 2. Parameterize your problem using statistical distributions
# Step 3. Write model in PyMC3, and justify your model structure

with pm.Model() as error_rate_model: 
    
    
    mu_distribution = pm.Uniform('mu_distribution', 20, 45)
    sigma_distribution = pm.Uniform('sigma_distribution', 0, 15)
    like = pm.Normal('likelihood', mu=mu_distribution, sigma=sigma_distribution,
                        observed=error_rate + error_rate_2 + error_rate_3)

    # Step 4. Inference
    step = pm.Metropolis()
    error_rate_trace = pm.sample(1000, step=step,cores=4)
```

```python
# Step 5. Interpret the results
pm.traceplot(error_rate_trace)
plt.show()
```

```python
pm.plot_posterior(error_rate_trace, 
                  color='#87ceeb', point_estimate='mean', 
                  rope=[30.9, 31.1], ref_val=31)
plt.show()
```
